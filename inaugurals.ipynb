```python
import numpy as np
import pandas as pd
import nltk
```

```python
from nltk.corpus import inaugural
print(inaugural.fileids())
```

```python
len(inaugural.raw(inaugural.fileids()[0]))
```

```python
inaug = pd.DataFrame(columns=['year','name','length','words','sents'])

for fileid in inaugural.fileids():
    
    #word characteristics
    length = len(inaugural.raw(fileid))
    words = len(inaugural.words(fileid))
    sents = len(inaugural.sents(fileid))
    vocab = len(set(word.lower() for word in inaugural.words(fileid)))
    
    #append word characteristics
    inaug = inaug.append({'year':fileid[:4], 'name':fileid[5:-4],
                          'length':length, 'words':words, 'sents':sents, 'vocab':vocab},
                         ignore_index=True)
inaug
```

```python
inaug.dtypes
```

```python
def convert_all_numcol_to_int(df):
    for col in df.columns:
        try: df[col] = df[col].astype(int)
        except: continue
    return(df)
```

```python
inaug = convert_all_numcol_to_int(inaug)
inaug
```

```python
inaug['awl'] = inaug.length/inaug.words
inaug['asl'] = inaug.length/inaug.sents
inaug['awps'] = inaug.words/inaug.sents
inaug['avps'] = inaug.vocab/inaug.sents
inaug['awpv'] = inaug.words/inaug.vocab
inaug
```

```python
inaug.index
```

```python
inaug['in_number'] = ''
president = 0
for i in inaug.index[1:]:
    if inaug['name'][i]!=inaug['name'][i-1]:
        president+=1
        inaug['in_number'][i] = president
        
    else:
        inaug['in_number'][i] = president
        inaug['in_number'][i-1] = president
        inaug['year'][i] = np.nan
```

```python
inaug
```

```python
inaug = inaug.groupby(['in_number'], as_index=False).mean()
```

### Get IQ Data

```python
iqs = pd.read_excel('presidents_iqs.xlsx', sheetname='data')
iqs.columns = [col.lower().replace(' ','_') for col in iqs.columns]
iqs = iqs.rename(columns={'iq_(corrected_for_data_reliability,_ages_18–26)[9]':'iq1',
                          'other_iq_estimates':'iq2','unnamed:_6':'iq3','name':'full_name'})
iqs.head()
```

```python
iqs.dtypes
```

```python
#iqs['Time in Office'] = iqs['Time in Office'].astype(str)
#iqs['time_in_office'] = iqs['time_in_office'].astype(str).apply(lambda s: (s.split('–'))[0])
iqs['office_start'] = iqs['time_in_office'].astype(str).apply(lambda s: (s.split('–'))[0])
iqs['office_end'] = iqs['time_in_office'].astype(str).apply(lambda s: (s.split('–'))[-1])
iqs = iqs.replace('present',np.nan)
iqs
```

```python
#Get average iq
iqs['iq'] = iqs[['iq1','iq2','iq3']].mean(axis=1)
```

```python
col_to_drop = ['political_party','lifespan','time_in_office','iq1','iq2','iq3']
for col in col_to_drop:
    iqs = iqs.drop(col,1)
iqs
```

```python
iqs = iqs.replace('van Buren', 'VanBuren')
iqs['name'] = iqs.full_name.astype(str).apply(lambda s: (s.split())[-1])
iqs
```

```python
convert_all_numcol_to_int(iqs)
```

### Merge inaug with iqs

```python
inaug_iqs = pd.merge(inaug, iqs, left_on='year', right_on='office_start', how='left')
inaug_iqs
```
